# En busca de los mejores 200 lugares para abrir pozos nuevos de petroleo para la compañia OilyGiant.

***Se trabajara con Aprendizaje supervisado, se utilizara el modelo de regresión lineal, el cual tiene como objectivo encontrar la mejor región en base a las caracteriscticas, se entrenara y probara el modelo.
Se complementara el modelo con un analisis de ganancias.***

```python
#Importación de librerias.

import pandas as pd
import numpy as np
from scipy import stats as st
from sklearn import set_config
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

set_config(print_changed_only=False)
```

## Entrenamiento y prueba de modelos para diferentes regiones.

###  Región 0

```python
#Carga de archivo e información de los datos Región 0

geo_0 = pd.read_csv('/content/drive/MyDrive/datasets/oilygiant/geo_data_0.csv')
geo_0.info()
```

```python
# Vizualiación aleatoria de los datos región 0.

geo_0.sample(3)
```

```python
# Estadisticas generales de los datos.

geo_0.describe()
```

```python
# División de los datos region 0.

features_0 = geo_0.drop(['id', 'product'], axis= 1)
target_0 = geo_0['product']
```

```python
# Segmentación de los datos región 0.

features_train_0, features_valid_0, target_train_0, target_valid_0 = train_test_split(features_0, target_0, test_size=0.25, random_state=12345)
```

```python
# Calculo de intervalo de confianza.

media_geo_0 = target_0.mean()
print("Media :" , media_geo_0)

confidence_interval_0 = st.t.interval(
    0.95, len(target_0)-1, target_0.mean(), target_0.sem())

print('Intervalo de confianza del 95 %:', confidence_interval_0)
```

```python
# Codificación de caracteristicas región 0.

scaler = StandardScaler()
scaler.fit(features_train_0)
features_train_0 = scaler.transform(features_train_0)
features_valid_0 = scaler.transform(features_valid_0)
```

```python
# Entrenamiento del modelo Regresion Lineal para la region 0

model =  LinearRegression()
model.fit(features_train_0, target_train_0)
predicted_valid_0 = model.predict(features_valid_0)
mse_0 = mean_squared_error(target_valid_0, predicted_valid_0)

print("MSE del modelo de regresión lineal en el conjunto de validación:", mse_0)
print("RMSE del modelo de regresión lineal en el conjunto de validación:", mse_0 ** 0.5)
```

## Calculo de ganancias

```python
# Comparativa de medias con la produción minima requerida de 111.10 barriles.

target = 111.1
print("Compartiva Región 0:", media_geo_0 - target)
print('-'*40)
print("Compartiva Región 1:", media_geo_1 - target)
print('-'*40)
print("Compartiva Región 2:", media_geo_2 - target)
```

# Calculo de ganancias para un conjunto de pozos.

```python
#Conversión de array a Series para usar los atributos de pandas.

predicted_valid_0 = pd.Series(predicted_valid_0)
predicted_valid_1 = pd.Series(predicted_valid_1)
predicted_valid_2 = pd.Series(predicted_valid_2)
```

### Región 0

```python
# creamos un data frame con los datos reales y predicciones de la región 0
# reinicie los índices para evitar problemas al crear el data frame

df_region_0 = pd.DataFrame({'reales':target_valid_0.reset_index(drop=True),
                            'prediccion':predicted_valid_0.reset_index(drop=True)})

# Iniciamos el ciclo for con 10000 iteraciones, este es el bootstrap, definimos la lista de las ganancias

ganancias_lista = []

for _ in range(10000):
    # 1. Elegimos 500 pozos aleatorios
    sample_500 = df_region_0.sample(n=500)

    # 2. Ordenamos de mayor a menor con respecto de la predicción

    order_sample_500 = sample_500.sort_values(by='prediccion', ascending=False)

    # 3. Tomamos los 200 mejores pozos segun su prediccion

    top_200 = order_sample_500.head(200)

    # Calculamos las ganancias con los datos reales,
    # esto es igual a la suma de petroleo multiplicado por 4500 (costo por unidad)
    # y restamos 100 millones (el costo de los 200 pozos)

    ganancia = np.sum(top_200['reales'])*4500 - 100000000
    ganancias_lista.append(ganancia)

# Ahora vemos el porcentaje de datos menor a 0 en la lista de ganancias ese es el riesgo de pérdida
porcentaje_perdida_reg_0 = 100 * (np.sum(np.array(ganancias_lista) < 0)/10000)
print(f'El porcentaje de perdida en la region 0 es de: {porcentaje_perdida_reg_0 }%')
```

### Región 1

```python
df_region_1 = pd.DataFrame({'reales':target_valid_1.reset_index(drop=True),
                            'prediccion':predicted_valid_1.reset_index(drop=True)})

ganancias_r1_lista = []

for _ in range(10000):
    sample_500_r1 = df_region_1.sample(n=500)
    order_sample_500_r1 = sample_500_r1.sort_values(by='prediccion', ascending=False)
    top_200_r1 = order_sample_500_r1.head(200)
    ganancia_r1 = np.sum(top_200_r1['reales'])*4500 - 100000000
    ganancias_r1_lista.append(ganancia_r1)

porcentaje_perdida_reg_1 = 100 * (np.sum(np.array(ganancias_r1_lista) < 0)/10000)
print(f'El porcentaje de perdida en la region 1 es de: {porcentaje_perdida_reg_1 }%')
```

### Región 2

```python
df_region_2 = pd.DataFrame({'reales':target_valid_2.reset_index(drop=True),
                            'prediccion':predicted_valid_2.reset_index(drop=True)})

ganancias_r2_lista = []

for _ in range(10000):
    sample_500_r2 = df_region_2.sample(n=500)
    order_sample_500_r2 = sample_500_r2.sort_values(by='prediccion', ascending=False)
    top_200_r2 = order_sample_500_r2.head(200)
    ganancia_r2 = np.sum(top_200_r2['reales'])*4500 - 100000000
    ganancias_r2_lista.append(ganancia_r2)

porcentaje_perdida_reg_2 = 100 * (np.sum(np.array(ganancias_r2_lista) < 0)/10000)
print(f'El porcentaje de perdida en la region 2 es de: {porcentaje_perdida_reg_2 }%')
```

En la región 1 y 2 se ve un riesgo inexitente en el calculo de las ganancias, sin embargo el entrenamiento y la prueba del modelo de Regresión Lineal, nos da evidencia que la región 1 representa una mejor opción para el desarrollo de pozos petrolíferos ya que su RMSE es menor, muy cercano a cero, lo cual nos dice que existe una pequeña diferencia entre el valor predicho y el valor real, a diferencia de las otras dos regiones.
